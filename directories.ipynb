```python
rootdir = r'\\fileserver\pim-documents\AAPIM DEAL DOWNLOAD\ADMIN SETTLEMENT FILES\BESA MTM Files'

def listdirs(rootdir):
    for file in os.listdir(rootdir):
        d = os.path.join(rootdir, file)
        if os.path.isdir(d):
            print(d)
            listdirs(d)
 
listdirs(rootdir)
```
